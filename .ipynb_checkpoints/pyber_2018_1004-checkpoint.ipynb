```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
# city_data_to_load = "data/city_data.csv"
# ride_data_to_load = "data/ride_data.csv"


# Display the data table for preview
```

```python
# Read the City and Ride Data
city_df = pd.read_csv("data/city_data.csv")
Ride_df = pd.read_csv("data/ride_data.csv")
```

```python
# Combine the data into a single dataset
city_rides_df = pd.merge(Ride_df,city_df,how='inner',on='city')
city_rides_df.head()
```

```python
#Driver count, Rides per city, Avg Fare and City type
rides_per_city = city_rides_df.groupby(['city']).count()["ride_id"]
avg_fare_city = city_rides_df.groupby(['city']).mean()["fare"]
driver_count =  city_rides_df.groupby(['city']).mean()["driver_count"]
city_type = city_df.set_index('city')['type']
```

```python
City_df = pd.DataFrame({ "Total Rides":rides_per_city,
                         "Average Fare":avg_fare_city,
                         "# of Drivers":driver_count,
                       "City Type":city_type
                       })
City_df.reset_index()
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types

# Build the scatter plots for each city types

# Incorporate the other graph properties

# Create a legend

# Incorporate a text label regarding circle size

# Save Figure
```

```python
# Obtain the x and y coordinates for each of the three city types
#City Type dfs(Urban, Suburban, Rural)
Urban = City_df.loc[city_type =='Urban']
Urban.head()
```

```python
Suburban = City_df.loc[city_type == 'Suburban']
Rural = City_df.loc[city_type =='Rural']
```

```python
# Build the scatter plots for each city types
#Urban scatter
plt.scatter(Urban['Total Rides'], Urban['Average Fare'], color='Gold', s = Urban['# of Drivers']*10, 
            edgecolor = 'black', label = 'Urban', alpha = .75)

#Suburban scatter
plt.scatter(Suburban['Total Rides'], Suburban['Average Fare'], color='LightSkyBlue',s = Suburban['# of Drivers']*10, 
            edgecolor = 'black', label = 'Suburban', alpha = .75)

#Rural scatter
plt.scatter(Rural['Total Rides'], Rural['Average Fare'], color='LightCoral',s = Rural['# of Drivers']*10, 
            edgecolor = 'black', label = 'Rural', alpha = .75)

# Incorporate the other graph properties
#Add title, x axis label and y axis label.
plt.title("Average Fare vs. Total Rides for City Types and Cities")
plt.xlabel("Total Rides")
plt.ylabel("Average Fare")


# Create a legend
#Add the legend.
plt.legend(loc= "best")

# Incorporate a text label regarding circle size
plt.text(40, 50,"Note: Circle size correlates with driver count per city.")

```

```python
# Show plot
plt.show()
```

## Total Fares by City Type

```python
# Calculate Type Percents
City_Type_Fares = city_rides_df.groupby(['type'])['fare'].sum()
City_Type_Fares
```

```python
# Build Pie Chart
# Build Pie Chart
colors = ["gold", "lightskyblue","lightcoral"]
explode = (0, 0, 0.2)

plt.title("% of Total Fares by City Type")
plt.pie(City_Type_Fares, explode=explode, labels=City_Type_Fares.index, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=130)
plt.axis("equal")
```

```python
# Show Figure
plt.show()
```

## Total Rides by City Type

```python
# Calculate Ride Percents
City_Type_Rides = city_rides_df.groupby(['type'])['ride_id'].count()
City_Type_Rides
# Build Pie Chart

# Save Figure
```

```python
# Build Pie Chart
colors = ["gold", "lightskyblue","lightcoral"]
explode = (0, 0, 0.2)

plt.title("% of Total Rides by City Type")
plt.pie(City_Type_Rides, explode=explode, labels=City_Type_Rides.index, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=180)
plt.axis("equal")
```

```python
# Show Figure
plt.show()
```

## Total Drivers by City Type

```python
# Calculate Driver Percents
City_Type_Drivers = city_df.groupby(['type'])['driver_count'].sum()
City_Type_Drivers
# Build Pie Charts

# Save Figure
```

```python
# Build Pie Chart
colors = ["gold", "lightskyblue","lightcoral"]
explode = (0, 0, 0.2)

plt.title("% of Total Drivers by City Type")
plt.pie(City_Type_Drivers, explode=explode, labels=City_Type_Drivers.index, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=180)
plt.axis("equal")
```

```python
# Show Figure
plt.show()
```

3 Observable Trends:
1) Rural customers have the fewest rides but the rides are more expensive
2) Urban city type has the most drivers, total dollar value of fares and number of rides
